# Observed Trip Explorer
Inspect inferred trips, edge lists, and OD matrices generated by `styrstell-infer-trips` and `styrstell-observed-od`.

```python
import pandas as pd
from pathlib import Path

processed_dir = Path('data/processed')
models_dir = Path('data/models')

trips = pd.read_parquet(processed_dir / 'trips.parquet')
edge_list = pd.read_parquet(processed_dir / 'observed_edge_list.parquet')
od_matrix = pd.read_parquet(models_dir / 'observed_od_matrix.parquet')
travel_stats = pd.read_parquet(models_dir / 'observed_travel_time.parquet')

trips
```
